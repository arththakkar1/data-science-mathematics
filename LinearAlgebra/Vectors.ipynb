# Theory of Vectors

A **vector** is a mathematical object that has both magnitude and direction. Vectors are fundamental in linear algebra and are used to represent quantities such as force, velocity, and displacement.

## Representation
- **Geometric**: As arrows in space, where the length represents magnitude and the arrowhead indicates direction.
- **Algebraic**: As an ordered list of numbers (components), e.g., \( \vec{v} = [v_1, v_2, ..., v_n] \) in \( n \)-dimensional space.

**Example:**
- The vector \( \vec{a} = [3, 4] \) in 2D space has magnitude \( \sqrt{3^2 + 4^2} = 5 \) and points from the origin to the point (3, 4).

## Types of Vectors
- **Zero Vector**: A vector with all components zero. Example: \( \vec{0} = [0, 0, 0] \).
- **Unit Vector**: A vector with magnitude 1. Example: \( \vec{u} = [1, 0, 0] \) in 3D.
- **Position Vector**: Represents the position of a point relative to the origin. Example: The position vector of point (2, -1) is \( [2, -1] \).

## Vector Operations
- **Addition**: Combine two vectors to get a new vector. Example: \( [1, 2] + [3, 4] = [4, 6] \).
- **Scalar Multiplication**: Multiply a vector by a number (scalar), changing its magnitude but not direction. Example: \( 2 \times [3, 4] = [6, 8] \).
- **Dot Product**: Measures the similarity of direction between two vectors. Example: \( [1, 2] \cdot [3, 4] = 1 \times 3 + 2 \times 4 = 11 \).
- **Cross Product**: Produces a vector perpendicular to two given vectors (in 3D). Example: \( [1, 0, 0] \times [0, 1, 0] = [0, 0, 1] \).

## Applications
Vectors are widely used in physics, engineering, computer graphics, and many other fields to model and solve real-world problems involving direction and magnitude.

**Examples:**
- In physics, velocity and force are represented as vectors.
- In computer graphics, vectors are used to describe positions, directions, and transformations.
- In navigation, vectors help determine displacement and direction between locations.
