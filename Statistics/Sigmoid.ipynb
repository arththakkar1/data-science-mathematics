# Sigmoid Function: Theory

The **sigmoid function** is a mathematical function that produces an S-shaped curve. It is widely used in statistics, machine learning, and neural networks, especially for classification problems.

## Definition

The most common form is the logistic sigmoid:

$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

where:
- $x$ is the input value
- $e$ is the base of the natural logarithm

## Properties

- The output ranges between 0 and 1.
- The curve is smooth and differentiable.
- For large positive $x$, $\sigma(x)$ approaches 1; for large negative $x$, $\sigma(x)$ approaches 0.
- The function is symmetric around $x = 0$.

## Applications in Data Science

- **Logistic Regression:** The sigmoid function maps predicted values to probabilities for binary classification.
- **Neural Networks:** Used as an activation function to introduce non-linearity.
- **Probability Theory:** Converts linear values to probabilities.

## Example

In logistic regression, the sigmoid function transforms the linear combination of features into a probability between 0 and 1, allowing us to interpret the output as the likelihood of a class.

## Note

The sigmoid function can suffer from vanishing gradients for extreme values, which is a consideration in deep learning.

```python
import pandas as pd

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

x = np.linspace(-10, 10, 100)

# get sigmoid output
y = sigmoid(x)

# get derivative of sigmoid
d = d_sigmoid(x)

df = pd.DataFrame({"x": x, "sigmoid(x)": y, "d_sigmoid(x)": d})

df
```
