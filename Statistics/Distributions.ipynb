# Theory of Distributions

In statistics, a **distribution** describes how the values of a random variable are spread or dispersed. It provides a mathematical function or graphical representation that shows all possible values and their associated probabilities.

## Types of Distributions
- **Discrete Distributions**: Concern variables that take on distinct, separate values (e.g., Binomial, Poisson).
- **Continuous Distributions**: Concern variables that can take on any value within a range (e.g., Normal, Exponential).

## Key Concepts
- **Probability Distribution**: Assigns probabilities to each possible value of a random variable.
- **Probability Density Function (PDF)**: Used for continuous variables to describe the likelihood of a value.
- **Probability Mass Function (PMF)**: Used for discrete variables to describe the probability of each value.
- **Cumulative Distribution Function (CDF)**: Gives the probability that a variable is less than or equal to a certain value.

## Importance
Distributions are fundamental in statistics for modeling uncertainty, making predictions, and performing hypothesis testing. Understanding distributions helps in selecting appropriate statistical methods and interpreting data correctly.

## Binomial Distribution Chart Explanation

The Binomial distribution models the number of successes in a fixed number of independent trials, each with the same probability of success. The chart shows the probability of getting a certain number of successes (from 0 to 10) when the probability of success in each trial is 0.5.

### Example: Binomial Distribution
Suppose you flip a fair coin 10 times. What is the probability of getting exactly 5 heads? The Binomial distribution models this scenario, where each flip is an independent trial with a probability of success (head) of 0.5. The chart shows the probabilities for getting 0 to 10 heads in 10 flips.

```python
# Binomial Distribution Chart
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, n+1)
probs = binom.pmf(x, n, p)
plt.bar(x, probs)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

## Poisson Distribution Chart Explanation

The Poisson distribution models the probability of a given number of events occurring in a fixed interval of time or space, when these events happen independently and at a constant average rate. The chart displays the probability of observing 0 to 9 events when the average rate (mu) is 3.

### Example: Poisson Distribution
Imagine a call center receives an average of 3 calls per minute. What is the probability that exactly 5 calls will be received in a given minute? The Poisson distribution models this scenario, and the chart shows the probabilities for receiving 0 to 9 calls per minute when the average rate is 3.

```python
# Poisson Distribution Chart
from scipy.stats import poisson

mu = 3
x = np.arange(0, 10)
probs = poisson.pmf(x, mu)
plt.bar(x, probs)
plt.title('Poisson Distribution (mu=3)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

## Normal Distribution Chart Explanation

The Normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric about the mean. It describes many natural phenomena. The chart shows the familiar bell-shaped curve for a mean (mu) of 0 and standard deviation (sigma) of 1.

### Example: Normal Distribution
Suppose the heights of adult men in a population are normally distributed with a mean of 0 and a standard deviation of 1 (after standardization). The chart shows the probability density for different heights, with most values clustering around the mean and fewer values farther from the mean.

```python
# Normal Distribution Chart
from scipy.stats import norm

mu, sigma = 0, 1
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf)
plt.title('Normal Distribution (mu=0, sigma=1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

## Exponential Distribution Chart Explanation

The Exponential distribution is a continuous distribution that describes the time between events in a Poisson process, where events occur continuously and independently at a constant average rate. The chart shows the probability density for different values, with a scale parameter of 1.

### Example: Exponential Distribution
Suppose the average time between arrivals of buses at a stop is 1 minute. What is the probability that the next bus arrives after 2 minutes? The Exponential distribution models the time between events in a Poisson process, and the chart shows the probability density for different waiting times.

```python
# Exponential Distribution Chart
from scipy.stats import expon

scale = 1
x = np.linspace(0, 8, 100)
pdf = expon.pdf(x, scale=scale)
plt.plot(x, pdf)
plt.title('Exponential Distribution (scale=1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
