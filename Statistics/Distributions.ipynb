# Theory of Distributions

In statistics, a **distribution** describes how the values of a random variable are spread or dispersed. Distributions are fundamental for understanding probability, data analysis, and statistical inference.

## Key Properties of Distributions
- **Mean**: The average or expected value of the distribution.
- **Median**: The middle value when data is ordered.
- **Mode**: The most frequently occurring value(s).
- **Variance and Standard Deviation**: Measures of how spread out the values are.
- **Skewness**: Describes the asymmetry of the distribution.
- **Kurtosis**: Describes the "tailedness" or extremity of outliers in the distribution.

## Why Study Distributions?
- Distributions help us model real-world phenomena and make predictions.
- They are used to calculate probabilities and critical values in hypothesis testing.
- Understanding the shape and properties of a distribution guides the choice of statistical methods.

## Types of Distributions

### Normal Distribution
The **Normal distribution** (or Gaussian distribution) is a symmetric, bell-shaped distribution where most values cluster around the mean. It is common in natural and social phenomena, such as heights or test scores. The normal distribution is fully described by its mean and standard deviation.

---

### Skewed Distribution
A **skewed distribution** is one where values are not symmetrically distributed. In a right-skewed (positively skewed) distribution, the tail is longer on the right, indicating more high-value outliers. In a left-skewed (negatively skewed) distribution, the tail is longer on the left, indicating more low-value outliers.

---

### Uniform Distribution
The **Uniform distribution** has all values equally likely within a certain range. The probability density is constant across the range. This distribution is often used to model random sampling and simulations.

---

### Bimodal Distribution
A **bimodal distribution** has two distinct peaks or modes. This can occur when data is drawn from two different populations or processes. Bimodal distributions often indicate the presence of subgroups within the data.

---

### Multimodal Distribution
A **multimodal distribution** has more than two peaks. This can indicate the presence of multiple groups or processes in the data, and suggests that the data may be a mixture of several different distributions.

---

## Applications
Distributions are used in fields such as statistics, data science, engineering, economics, and the natural and social sciences. They are essential for:
- Modeling uncertainty
- Simulating random processes
- Making inferences from data
- Designing experiments and surveys
- Quality control and risk assessment

```python
# Normal Distribution Chart
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu, sigma = 0, 1
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf)
plt.title('Normal Distribution (mu=0, sigma=1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Right-Skewed Distribution Chart
from scipy.stats import skewnorm

x = np.linspace(-4, 8, 200)
skewed_pdf = skewnorm.pdf(x, a=6, loc=2, scale=2)
plt.plot(x, skewed_pdf)
plt.title('Right-Skewed Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Uniform Distribution Chart
from scipy.stats import uniform

x = np.linspace(0, 1, 100)
pdf = uniform.pdf(x, loc=0, scale=1)
plt.plot(x, pdf)
plt.title('Uniform Distribution (0, 1)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Bimodal Distribution Chart
x = np.linspace(-6, 10, 300)
pdf = 0.5 * norm.pdf(x, loc=-1, scale=1) + 0.5 * norm.pdf(x, loc=4, scale=1)
plt.plot(x, pdf)
plt.title('Bimodal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Multimodal Distribution Chart
x = np.linspace(-8, 12, 400)
pdf = (0.3 * norm.pdf(x, loc=-3, scale=1) +
       0.4 * norm.pdf(x, loc=2, scale=1) +
       0.3 * norm.pdf(x, loc=7, scale=1))
plt.plot(x, pdf)
plt.title('Multimodal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
