# Probability

## Theory of Probability

Probability is a branch of mathematics that quantifies the likelihood of events occurring. It is fundamental to statistics, data science, and many scientific disciplines, helping us make predictions and informed decisions under uncertainty.

### Definition
Probability measures how likely an event is to happen, expressed as a number between 0 and 1:
- 0 means the event is impossible.
- 1 means the event is certain.
- Values between 0 and 1 indicate varying degrees of likelihood.

### Formula
For an event $A$:
$$
P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}}
$$

### Properties
- The probability of any event is always between 0 and 1.
- The sum of probabilities of all possible outcomes in a sample space is 1.
- Complement rule: $P(\text{not }A) = 1 - P(A)$

### Example
If you roll a fair six-sided die, the probability of getting a 4 is:
$$
P(4) = \frac{1}{6}
$$

Probability is used to model random phenomena, assess risk, and guide decision-making in data science and beyond.

## More Probability Examples

### Example 1: Coin Toss
If you toss a fair coin, the probability of getting heads is:
$$
P(\text{Heads}) = \frac{1}{2}
$$

### Example 2: Drawing a Card
If you draw a card from a standard deck of 52 cards, the probability of getting an Ace is:
$$
P(\text{Ace}) = \frac{4}{52} = \frac{1}{13}
$$

### Example 3: Rolling Two Dice
The probability of getting a sum of 7 when rolling two fair six-sided dice:
- Possible pairs: (1,6), (2,5), (3,4), (4,3), (5,2), (6,1) â€” 6 outcomes
- Total outcomes: $6 \times 6 = 36$
$$
P(\text{Sum of 7}) = \frac{6}{36} = \frac{1}{6}
$$

### Example 4: Probability of Not Occurring
If the probability of rain tomorrow is 0.3, the probability that it does not rain is:
$$
P(\text{No Rain}) = 1 - 0.3 = 0.7
$$

### Example 5: Multiple Events
If you randomly select a student from a class of 20, and 5 are left-handed, the probability of selecting a left-handed student is:
$$
P(\text{Left-handed}) = \frac{5}{20} = 0.25
$$

## Probability Using Combinations (nCr)

In probability, combinations (denoted as $nCr$ or $C(n, r)$) are used to count the number of ways to choose $r$ items from $n$ distinct items without regard to order. This is essential for calculating probabilities in scenarios where the arrangement does not matter.

### Formula for Combinations
$$
C(n, r) = \frac{n!}{r!(n - r)!}
$$
where $n!$ is the factorial of $n$.

### Using nCr in Probability
When calculating the probability of an event involving combinations:
$$
P(A) = \frac{\text{Number of favorable combinations}}{\text{Total number of possible combinations}}
$$

### Example
Suppose you have a deck of 52 cards and want to find the probability of drawing 4 aces when picking 4 cards:
- Total ways to choose 4 cards: $C(52, 4)$
- Ways to choose 4 aces: $C(4, 4) = 1$

$$
P(\text{4 aces}) = \frac{1}{C(52, 4)}
$$

Combinations are widely used in probability problems involving selection, such as lottery, card games, and sampling.

## More Examples of Probability Using Combinations (nCr)

### Example 1: Lottery Selection
A lottery requires you to choose 6 numbers out of 49. How many possible combinations are there?
$$
C(49, 6) = \frac{49!}{6! \times 43!} = 13,983,816
$$

### Example 2: Committee Formation
From a group of 10 people, how many ways can you form a committee of 4?
$$
C(10, 4) = \frac{10!}{4! \times 6!} = 210
$$

### Example 3: Drawing Red Balls
Suppose a box contains 8 red and 7 blue balls. If you randomly select 5 balls, what is the probability that exactly 3 are red?
- Ways to choose 3 red: $C(8, 3) = 56$
- Ways to choose 2 blue: $C(7, 2) = 21$
- Total ways to choose 5 balls: $C(15, 5) = 3003$

$$
P(\text{3 red, 2 blue}) = \frac{56 \times 21}{3003} \approx 0.392
$$

### Example 4: Selecting Students
From a class of 12 students, how many ways can you select a president, vice-president, and secretary (order matters)?
This is a permutation, not a combination:
$$
P(12, 3) = \frac{12!}{(12-3)!} = 1320
$$

### Example 5: Card Hands
How many ways can you choose 5 cards from a deck of 52?
$$
C(52, 5) = 2,598,960
$$
