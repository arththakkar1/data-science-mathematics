# Range

## Theory of Range

The range is a simple measure of dispersion in a dataset. It represents the difference between the largest and smallest values, giving an idea of how spread out the data is.

### How to Calculate the Range
- Identify the maximum value in the dataset.
- Identify the minimum value in the dataset.
- Subtract the minimum value from the maximum value:

$$
\text{Range} = \text{Maximum Value} - \text{Minimum Value}
$$

### Properties
- The range is easy to calculate and understand.
- It is highly sensitive to outliers, as it only considers the two extreme values.
- The range does not provide information about the distribution of values between the extremes.

### Example
For the dataset [4, 7, 9, 15, 21], the range is $21 - 4 = 17$.

The range is useful for quickly assessing the spread of data, but for more detailed analysis, other measures like variance or standard deviation are often preferred.

```python
import numpy as np
Data = np.array([2, 4, 5, 4, 6, 4, 7, 5])
data_range = np.max(Data) - np.min(Data)
print("Range:", data_range)
```
